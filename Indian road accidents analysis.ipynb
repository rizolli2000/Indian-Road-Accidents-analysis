```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("India_Injury_Road_Accident_Fatality_2017-2020.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(axis=0,how='any',inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.columns=[i.replace(' ','_')for i in df.columns]
df.columns
```

```python
df.sort_values(by='Road_Accidents__during_2018',ascending=False)
```

```python
highest_acc_2018=df.sort_values('Road_Accidents__during_2018',ascending=False)[1:21]
highest_acc_2018
```

```python
plt.figure(figsize=(15,9))
sns.barplot(data=highest_acc_2018,x='State',y='Road_Accidents__during_2018')
plt.title("Highest Accident in 2018 State Wise",fontsize=14,pad=14)
plt.xlabel('State',fontsize=12)
plt.ylabel('Accidents During 2018',fontsize=12)

plt.tick_params(axis='x',which='major',labelsize=12,rotation=90)
plt.tick_params(axis='y',which='major',labelsize=12)
```

```python
low_acc_2018=df.sort_values(by='Road_Accidents__during_2018')[:5]
low_acc_2018
```
